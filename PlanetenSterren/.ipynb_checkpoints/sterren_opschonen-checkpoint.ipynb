```python
import pandas as pd
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
tab = pd.read_csv('SterrenCoordinatenVolledig.csv',
                  sep = ',',
                  dtype = {'pl_name' : 'string',
                           'hostname' : 'string',
                           'ra' : 'float',
                           'dec' : 'float',
                           'sy_dist' : 'float',
                           'sy_snum' : 'int',
                           'sy_pnum' : 'int',
                           'st_spectype' : 'string',
                           'st_teff' : 'float',
                           'st_rad' : 'float',
                           'st_mass' : 'float',
                           'rastr' : 'string',
                           'decstr' : 'string'
                          },
                 header = 19)
```

```python
tab.drop(['rastr', 'decstr'], axis=1, inplace=True)
tab.dropna(inplace=True)
tab.drop_duplicates(subset=['hostname'], inplace=True)

tab.rename(columns={'pl_name' : 'planet_name',
                    'hostname' : 'starname',
                    'ra' : 'right_ascencion',
                    'dec' : 'declination',
                    'sy_dist' : 'distance',
                    'sy_snum' : 'number_stars',
                    'sy_pnum' : 'number_planets',
                    'st_spectype' : 'spectral_type',
                    'st_teff' : 'star_temp',
                    'st_rad' : 'star_radius',
                    'st_mass' : 'star_mass'},
           inplace=True)
```

```python
tab['alpha'] = tab['right_ascencion'].apply(math.radians)
tab['delta'] = tab['declination'].apply(math.radians)
```

```python
tab['X'] = tab['delta'].apply(math.cos) * tab['alpha'].apply(math.cos) * tab['distance']
tab['Y'] = tab['delta'].apply(math.cos) * tab['alpha'].apply(math.sin) * tab['distance']
tab['Z'] = tab['delta'].apply(math.sin) * tab['distance']
```

```python
tab.to_csv('sterren_zonder_na.csv')
```

```python
fig = plt.figure(figsize=(10,10))
ax = Axes3D(fig, auto_add_to_figure=False)
ax.set_xlim3d(-600, 600)
ax.set_ylim3d(-600, 600)
ax.set_zlim3d(-600, 600)
fig.add_axes(ax)
ax.scatter(tab.X, tab.Y, tab.Z, c=tab.star_mass)
```
